```python
import argparse
import time
import itertools

import numpy as np
import pandas as pd

name  = "data_"
nodes = ["5653", "55ad", "55e4", "5599", "55dd", "5565", "560b", "5632", "55b3", "5563", "630a"]
#nodes = ["5653"]
ext = '.csv'

names  = ["FSK", "OQPSK", "OFDM"] 
values = ["fsk_rssi", "oqpsk_rssi", "ofdm_rssi"]

total_packets_node_mod = 99 * 24 * 60

total_rx_packets = 0
pdr_values = {x: [] for x in names} 
pdr_totals = {x: [] for x in ["fsk_1", "fsk_2", "fsk_3","oqpsk_1","oqpsk_2","oqpsk_3","ofdm_1","ofdm_2","ofdm_3"]}

# Iterate over all nodes
for node in nodes:
    print("Parsing node = {} ".format(node))
    
    # Create measurement name
    filename = name + node + ext
    
    # Read values from csv
    filename = name + node + ext
    data = pd.read_csv(filename, index_col=0) 
    
    fsk_1   = 100 * data[data['pkt_retry']==0]['fsk_rssi'].count() / total_packets_node_mod
    oqpsk_1 = 100 * data[data['pkt_retry']==0]['oqpsk_rssi'].count() / total_packets_node_mod
    ofdm_1  = 100 * data[data['pkt_retry']==0]['ofdm_rssi'].count() / total_packets_node_mod
    
    fsk_2   = 100 * data[data['pkt_retry']==1]['fsk_rssi'].count() / total_packets_node_mod
    oqpsk_2 = 100 * data[data['pkt_retry']==1]['oqpsk_rssi'].count() / total_packets_node_mod
    ofdm_2  = 100 * data[data['pkt_retry']==1]['ofdm_rssi'].count() / total_packets_node_mod
    
    fsk_3   = 100 * data[data['pkt_retry']==2]['fsk_rssi'].count() / total_packets_node_mod
    oqpsk_3 = 100 * data[data['pkt_retry']==2]['oqpsk_rssi'].count() / total_packets_node_mod
    ofdm_3  = 100 * data[data['pkt_retry']==2]['ofdm_rssi'].count() / total_packets_node_mod
    
    pdr_totals["fsk_1"].append(fsk_1)
    pdr_totals["fsk_2"].append(fsk_2)
    pdr_totals["fsk_3"].append(fsk_3)
    pdr_totals["oqpsk_1"].append(oqpsk_1)
    pdr_totals["oqpsk_2"].append(oqpsk_2)
    pdr_totals["oqpsk_3"].append(oqpsk_3)
    pdr_totals["ofdm_1"].append(ofdm_1)
    pdr_totals["ofdm_2"].append(ofdm_2)
    pdr_totals["ofdm_3"].append(ofdm_3)
    
    print("{:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f}".format(fsk_1, fsk_2, fsk_3, oqpsk_1, oqpsk_2, oqpsk_3, ofdm_1, ofdm_2, ofdm_3))

    
print("*****")
print("{:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f}".format(
      np.mean(pdr_totals["fsk_1"]), np.mean(pdr_totals["fsk_2"]), np.mean(pdr_totals["fsk_3"]),
      np.mean(pdr_totals["oqpsk_1"]), np.mean(pdr_totals["oqpsk_2"]), np.mean(pdr_totals["oqpsk_3"]),
      np.mean(pdr_totals["ofdm_1"]), np.mean(pdr_totals["ofdm_2"]), np.mean(pdr_totals["ofdm_3"])))
print("{:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f} & {:.1f}".format(
      np.std(pdr_totals["fsk_1"]), np.std(pdr_totals["fsk_2"]), np.std(pdr_totals["fsk_3"]),
      np.std(pdr_totals["oqpsk_1"]), np.std(pdr_totals["oqpsk_2"]), np.std(pdr_totals["oqpsk_3"]),
      np.std(pdr_totals["ofdm_1"]), np.std(pdr_totals["ofdm_2"]), np.std(pdr_totals["ofdm_3"])))
```
