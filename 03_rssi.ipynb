```python
import argparse
import time
import itertools

import numpy as np
import pandas as pd

name  = "data_"
nodes = ["5653", "55ad", "55e4", "5599", "55dd", "5565", "560b", "5632", "55b3", "5563", "630a"]
#nodes = ["5653"]
ext = '.csv'

names  = ["FSK", "OQPSK", "OFDM"] 
values = ["fsk_rssi", "oqpsk_rssi", "ofdm_rssi"]

rssi_values = {x: [] for x in names} 

# Iterate over all nodes
for node in nodes:
    print("Parsing node = {} ".format(node))
    
    # Create measurement name
    filename = name + node + ext
    
    # Read values from csv
    filename = name + node + ext
    data = pd.read_csv(filename) 
    
    rssi_node = []
    for n, v in zip(names, values):
        rssi = data[v].mean()
        rssi_node.append(rssi)
        rssi_values[n].append(rssi)
        print(" {} - RSSI = {:.1f}".format(n, rssi))
    
    print(" MEAN RSSI {:.1f}".format(np.mean(rssi_node)))    
    print("*****")
    
for n in names:
    data = rssi_values[n]
    mean = np.mean(data)
    std = np.std(data)
    print("{} - MEAN = {:.1f}, STD = {:.1f}".format(n, mean, std))
```
