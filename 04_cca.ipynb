```python
import argparse
import time
import itertools

import numpy as np
import pandas as pd

name  = "data_"
nodes = ["5653", "55ad", "55e4", "5599", "55dd", "5565", "560b", "5632", "55b3", "5563", "630a"]
#nodes = ["5653"]
ext = '.csv'

names  = ["FSK", "OQPSK", "OFDM"] 
values = ["fsk_cca_rssi", "oqpsk_cca_rssi", "ofdm_cca_rssi"]

cca_values = {x: [] for x in names} 

# Iterate over all nodes
for node in nodes:
    print("Parsing node = {} ".format(node))
    
    # Create measurement name
    filename = name + node + ext
    
    # Read values from csv
    filename = name + node + ext
    data = pd.read_csv(filename)
    
    # Clip data
    data["fsk_cca_rssi"] = data["fsk_cca_rssi"].where(data["fsk_cca_rssi"] < -30, np.nan)
    data["oqpsk_cca_rssi"] = data["oqpsk_cca_rssi"].where(data["oqpsk_cca_rssi"] < -30, np.nan)
    data["ofdm_cca_rssi"] = data["ofdm_cca_rssi"].where(data["ofdm_cca_rssi"] < -30, np.nan)
    
    cca_node = []
    for n, v in zip(names, values):
        cca = data[v].mean()
        cca_node.append(cca)
        cca_values[n].append(cca)
        print(" {} - CCA = {:.1f}".format(n, cca))
    
    print(" MEAN CCA {:.1f}".format(np.mean(cca_node)))    
    print("*****")
    
for n in names:
    data = cca_values[n]
    mean = np.mean(data)
    std = np.std(data)
    print("{} - CCA = {:.1f}, STD = {:.1f}".format(n, mean, std))
```
